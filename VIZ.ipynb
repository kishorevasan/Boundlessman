```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./CLEANDATA.csv")
df = df.fillna(0)
df['year'] = df['year'].replace({'Autumn 1999': '1999','Autumn 2000':'2000','Spring 1999':'1999','Spring 2000':'2000'})
df.head()
```

```python
yearcounts = df.year.value_counts().to_frame()
yearcounts.index.name = 'idx'
yearcounts.reset_index(inplace = True)
yearcounts.columns = ['year','value']
yearcounts = yearcounts.sort_values('year')
yearcounts.head()
```

```python
plt.bar(yearcounts.year,yearcounts.value)
plt.xticks(rotation = 90)
plt.show()
```
